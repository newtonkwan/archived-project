```python
# Function 0: Convert an array into a unique string
def array_to_ID(array):
    '''
    Takes a 3x8 array and converts it into a unique ID 
    '''
    reshaped_array = array.reshape(1,12)
    ID = ''
    for square in reshaped_array[0]:
        square = tuple(square)
        if square == (0,0):
            ID += '0'
        if square == (1,0):
            ID += '1'
        if square == (0,1):
            ID += '2'
        if square == (2,0):
            ID += '3'
        if square == (0,2):
            ID += '4'
    
    return ID
```

```python
# Function 2: Convert a str into a tuple 
def num_to_tuple(number):
    '''
    Converts a real number into a tuple 
    '''
    if number == '0':
        return (0,0)
    if number == '1':
        return (1,0)
    if number == '2': 
        return (0,1)
    if number == '3':
        return (2,0)
    if number == '4':
        return (0,2)
    return 
```

```python
import string
digs = string.digits + string.ascii_letters
def int2base(x, base):
    if x < 0:
        sign = -1
    elif x == 0:
        return digs[0]
    else:
        sign = 1

    x *= sign
    digits = []

    while x:
        digits.append(digs[int(x % base)])
        x = int(x / base)

    if sign < 0:
        digits.append('-')

    digits.reverse()
    
    if len(digits) == 1:
        y = '00000000000'+ ''.join(digits)   
    if len(digits) == 2:
        y = '0000000000'+ ''.join(digits)
    if len(digits) == 3:
        y = '000000000'+ ''.join(digits)
    if len(digits) == 4:
        y = '00000000'+ ''.join(digits)
    if len(digits) == 5:
        y = '0000000'+ ''.join(digits)
    if len(digits) == 6:
        y = '000000'+ ''.join(digits)
    if len(digits) == 7:
        y = '00000'+ ''.join(digits)
    if len(digits) == 8:
        y = '0000'+ ''.join(digits)
    if len(digits) == 9:
        y = '000'+ ''.join(digits)
    if len(digits) == 10:
        y = '00'+ ''.join(digits)
    if len(digits) == 11:
        y = '0'+ ''.join(digits)

    return y
    
```
